```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Online Retail.csv')
df.info()
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
df.describe()
```

```python
df
```

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Date'] = df['InvoiceDate'].dt.date
df['Day'] = df['Date'].apply(lambda x : x.replace(month=1))
df['Amount'] = df['Quantity'] * df['UnitPrice']
```

```python
day_sales = df.groupby(['Day']).agg({'CustomerID':'nunique',
                                    'Quantity':'sum',
                                    'UnitPrice':'sum'})
day_sales.columns = ['TotalCus','TotalSales','TotalPrice']
day_sales.reset_index(inplace=True)
day_sales = day_sales.iloc[20:,:]
day_sales.head()
```

```python
for i in range(len(day_sales['Day'])):
    day_sales['Day'].values[i] = day_sales['Day'].values[i].day
```

```python
plt.figure(figsize=(5,12))
sns.barplot(x='Amount',y='Country',data=df,color="#564d4d", label = 'TotalCus')
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(x='Day',y='TotalPrice',data=day_sales,color="#e50914",linewidth=2.5)
```

```python
plt.figure(figsize=(12,5))
sns.lineplot(x='Day',y='TotalCus',data=day_sales,color="#831010",linewidth=2.5)
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x='Day',y='TotalSales',data=day_sales,color='#221f1f')
```

```python
product_sales = df.groupby(['Description'])[['Amount','Quantity']].sum()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='Quantity', y='Amount', color='#43465e', data=product_sales)
```

```python
fig, ax = plt.subplots(3,2, figsize=(18,12))
gs = ax[1,1].get_gridspec()
for a in ax[1:,-1]:
    a.remove()
axbig = fig.add_subplot(gs[1:, -1])

sns.lineplot(x='Day',y='TotalCus',data=day_sales,color="#831010",linewidth=2.5, ax = ax[0,0])
sns.lineplot(x='Day',y='TotalPrice',data=day_sales,color="#e50914",linewidth=2.5, ax = ax[1,0])
sns.barplot(x='Day',y='TotalSales',data=day_sales,color='#221f1f',ax = ax[2,0])
sns.scatterplot(x='Quantity', y='Amount', color='#43465e', data=product_sales, ax =ax[0,1])
sns.barplot(x='Amount',y='Country',data=df,color="#564d4d", label = 'TotalCus', ax= axbig)

ax[0,0].set_title('Customers in one month')
ax[1,0].set_title('Total Sales in one month')
ax[2,0].set_title('Total Sold Products in one month')
ax[0,1].set_title('Amount vs Quantity')
axbig.set_title('Country Sales')

plt.tight_layout(pad=3.0)
fig.suptitle('\nOnline Retail Dashboard', fontsize=20)
fig.subplots_adjust(top=0.88)
sns.despine(bottom = True, left = True)

plt.show()
```
